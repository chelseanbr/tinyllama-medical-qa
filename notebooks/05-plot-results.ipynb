```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(10, 8))

# Define the boxes and their labels
boxes = [
    "MedQuAD Dataset",
    "MCQ Generation with GPT (Phi-2 or GPT-3.5)",
    "Rejection Sampling: Filter Good MCQs",
    "Final MCQ Dataset",
    "Fine-Tuning TinyLLaMA (with LoRA)",
    "Evaluation on Hold-Out Test Set"
]

# Draw the boxes
for i, text in enumerate(boxes):
    y = 1 - i * 0.17
    rect = patches.FancyBboxPatch(
        (0.2, y), 0.6, 0.1,
        boxstyle="round,pad=0.02", 
        linewidth=2, edgecolor="black", facecolor="lightblue"
    )
    ax.add_patch(rect)
    ax.text(0.5, y + 0.05, text, ha="center", va="center", fontsize=11, weight="bold")

# Draw the arrows
for i in range(len(boxes) - 1):
    start_y = 1 - i * 0.17
    end_y = 1 - (i + 1) * 0.17
    ax.annotate(
        '', xy=(0.5, end_y + 0.09), xytext=(0.5, start_y - 0.01),
        arrowprops=dict(arrowstyle="->", lw=2)
    )

# Clean up plot
ax.axis("off")
plt.title("Workflow: Rejection-Sampling and Fine-Tuning for Layperson Medical QA", fontsize=14, weight="bold")
plt.show()
```
